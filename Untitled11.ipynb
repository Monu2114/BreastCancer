```python
#1a
import numpy as np
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/dataR2.csv')
for i in range(5):
  print(data.head(i))
```

```python
#1b
data.isnull().sum()
```

```python
#1c
data['Classification'] = data['Classification'].replace({'Healthy': 0, 'Patients': 1})
data.head(1)
```

```python
#1d
from sklearn.feature_selection import chi2
targetcol=data.drop(columns=['Classification'])
targetseries=data['Classification']
targetseries.head(0)
chi2_stat, p_values = chi2(targetcol, targetseries)
chi2_results = pd.DataFrame({'Feature': targetcol.columns, 'Chi-Square Statistic': chi2_stat, 'p-value': p_values})

# Display the results sorted by Chi-Square statistic in descending order
chi2_results = chi2_results.sort_values(by='Chi-Square Statistic', ascending=False)
print(chi2_results)
```

```python
least_important_feature = chi2_results['Feature'][chi2_results['Chi-Square Statistic'].idxmin()]
least_important_feature
```

```python
#Q2
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(targetcol, targetseries, test_size=0.2, random_state=42)
X_train
```

```python
#Q3
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
clf
```

```python
#Q4
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
y_pred=clf.predict(X_test)
confusion = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
```

```python
#Q5
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import log_loss, accuracy_score

# Create a Decision Tree classifier with entropy as the criterion
clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)

# Fit the classifier to the training data
clf_entropy.fit(X_train, y_train)

# Make predictions on the test set
y_pred_entropy = clf_entropy.predict(X_test)

# Calculate the log loss on the test set
log_loss_entropy = log_loss(y_test, y_pred_entropy)

# Calculate accuracy on the test set
accuracy_entropy = accuracy_score(y_test, y_pred_entropy)

# Print the results for the entropy criterion
print("Results for Entropy Criterion:")
print("Log Loss:", log_loss_entropy)
print("Accuracy:", accuracy_entropy)

# Now, you have evaluated the model using entropy criterion and log loss metric
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Create a Decision Tree classifier
clf = DecisionTreeClassifier(random_state=42)

# Define the hyperparameter grid to search
param_grid = {
    'max_depth': [10, 100],
    'min_samples_split': [4, 6, 8],
    'max_features': ["auto", "sqrt", "log2"]
}

# Create a grid search object
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Get the best model with the best hyperparameters
best_model = grid_search.best_estimator_

# Evaluate the best model on the test set
y_pred_best = best_model.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best)
print("Accuracy of the Best Model:", accuracy_best)
```
